```python
drop_column_names=[]

for column_name,missing_percnt in zip((df.isnull().sum().index),(df.isnull().sum()/df.shape[0]).values):
    print(column_name,missing_percnt*100)
    if missing_percnt>0.3:
        drop_column_names.append(column_name)
```

```python

df[drop_column_names].isnull().sum()/df.shape[0]
```

```python
null_report=df.isna().sum()/df.shape[0]
          
```

```python
  drop_column_names=null_report[null_report>0.3].index
```

```python
df.drop(list(drop_column_names),axis=1,inplace=True)
```
